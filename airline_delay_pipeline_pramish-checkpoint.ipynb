## Airline Delay - Data Pipeline

### Step 1 - Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### Step 2 - Extract (Load the CSV)

```python
df = pd.read_csv('Airline_Delay_Cause.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

### Step 3 - Transform (Clean + Prepare the data)

```python
# check for duplicates
df.duplicated().sum()
```

```python
# check missing values
df.isnull().sum()
```

```python
# dropping rows with nulls - the missing rows are very few compared to 171k rows so its fine
df.dropna(inplace=True)
print('rows after dropping nulls:', df.shape[0])
```

```python
# confirm no more nulls
df.isnull().sum().sum()
```

```python
# adding a total_delay_minutes column - sum of all delay causes
df['total_delay_minutes'] = df['carrier_delay'] + df['weather_delay'] + df['nas_delay'] + df['security_delay'] + df['late_aircraft_delay']
```

```python
# adding on_time_rate - what % of flights landed on time
df['on_time_rate'] = ((df['arr_flights'] - df['arr_del15']) / df['arr_flights']).round(3)
```

```python
df.head()
```

### Step 4 - Load (Save cleaned data)

```python
df.to_csv('airline_delay_cleaned.csv', index=False)
print('saved!')
```

### Visualizations

#### 1. Correlation Heatmap

```python
num_df = df[['arr_flights','arr_del15','carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay','arr_cancelled','on_time_rate']]

plt.figure(figsize=(10, 7))
sns.heatmap(num_df.corr(), annot=True, fmt='.2f', cmap='YlOrRd', linewidths=0.5)
plt.title('Correlation between delay-related columns')
plt.tight_layout()
plt.show()
```

**Insight:** `late_aircraft_delay` and `carrier_delay` are both highly correlated with `arr_del15` (total delayed flights). `on_time_rate` has a strong negative correlation with delays as expected. `security_delay` is barely correlated with anything - it's basically a non-factor.

#### 2. Distribution of arrival delays

```python
plt.figure(figsize=(9, 5))
sns.histplot(df['arr_delay'], bins=60, kde=True, color='steelblue')
plt.title('Distribution of Total Arrival Delay (minutes)')
plt.xlabel('Arrival Delay (minutes)')
plt.ylabel('Frequency')
plt.xlim(0, 50000)
plt.show()
```

**Insight:** The distribution is heavily right-skewed. Most airport-carrier combinations have low total delay minutes, but a small number have extremely high values. This makes sense - busy airports/carriers accumulate delay.

#### 3. Distribution of on-time rate

```python
plt.figure(figsize=(9, 5))
sns.histplot(df['on_time_rate'], bins=40, kde=True, color='seagreen')
plt.title('Distribution of On-Time Rate per Route')
plt.xlabel('On-Time Rate (0 = all delayed, 1 = all on time)')
plt.ylabel('Frequency')
plt.show()
```

**Insight:** Most routes have an on-time rate between 0.7 and 0.95 - meaning roughly 70-95% of flights land on time. The peak is around 0.80-0.85, so about 1 in 5 or 6 flights gets delayed on a typical route.

#### 4. Which delay cause hurts the most? (Bar chart)

```python
cause_totals = df[['carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay']].sum().sort_values(ascending=False)

plt.figure(figsize=(9, 5))
sns.barplot(x=cause_totals.index, y=cause_totals.values / 1e6, palette='Blues_d')
plt.title('Total Delay Minutes by Cause (all years combined)')
plt.xlabel('Delay Cause')
plt.ylabel('Total Minutes (millions)')
plt.xticks(rotation=20)
plt.show()
```

**Insight:** Late aircraft is the biggest single cause of delay - not weather, not the carrier. This is the "domino effect" - one late plane causes all its next flights to be late too. Carrier-related delays come second.

#### 5. Average on-time rate by month (are some months worse?)

```python
month_ontime = df.groupby('month')['on_time_rate'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=month_ontime, x='month', y='on_time_rate', marker='o', color='darkorange', linewidth=2)
plt.title('Average On-Time Rate by Month')
plt.xlabel('Month')
plt.ylabel('Avg On-Time Rate')
plt.xticks(range(1, 13), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

**Insight:** On-time rates dip in June and July (summer thunderstorms, peak traffic) and again in December (winter weather + holiday rush). September and October are clearly the best months to fly if you want to be on time.

#### 6. Top 10 carriers by average delay minutes per flight

```python
carrier_avg = df.groupby('carrier_name')['arr_delay'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=carrier_avg.values, y=carrier_avg.index, palette='Reds_r')
plt.title('Top 10 Carriers - Avg Arrival Delay per Entry')
plt.xlabel('Average Delay (minutes)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

**Insight:** The carriers at the top of this list aren't necessarily "worst" - they tend to be larger airlines flying more routes (more chances to accumulate delay). But it still shows which carriers passengers feel the most delays with.

---
### Key Takeaways

**1. Late aircraft is the real villain**  
Looking at graph 4, late aircraft delay accounts for the most total delay minutes across all years. Weather gets blamed publicly but it's actually less impactful overall.

**2. Summer and December are the worst months to fly**  
From graph 5, on-time rates clearly drop in June-July and December. If you want fewer delays, fly in September or October.

**3. Most routes are actually pretty reliable**  
The on-time rate distribution (graph 3) peaks around 0.80-0.85, meaning 80%+ of flights land on time on most routes. It's the extreme cases that drag the average down.

**4. Security delays are basically zero**  
The heatmap and bar chart both show security_delay is nearly irrelevant. TSA rarely causes systemic delay.
